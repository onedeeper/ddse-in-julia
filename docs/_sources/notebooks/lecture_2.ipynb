## The derivative

The *derivative* of a function $f(x)$: The rate of change of $f(x)$ with respect to an independent variable $x$

```julia
f(x) = x^2
out = []

for i in 1:20
    push!(out,f(i))
end
```

```julia
using Plots
plot(out, label ="f(x)")
```
